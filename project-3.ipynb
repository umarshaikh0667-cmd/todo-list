```python
class ToDoList:
    def __init__(self):
        self.tasks = []
        
    def add_task(self, task_name):
        new_task = {"name": task_name, "completed": False}
        self.tasks.append(new_task)
        print(f"Added: '{task_name}'")
        
    def mark_done(self, task_name):
        for task in self.tasks:
            if task["name"].lower() == task_name.lower():
                task["completed"] = True
                print(f"Awesome! Marked '{task_name}' as done.")
                return 
        
        print(f"Hmm, couldn't find '{task_name}' in your list.")
        
    def show_pending(self):
        print("\n--- Pending Tasks ---")
        pending_count = 0
        
        for task in self.tasks:
            if task["completed"] == False:
                print(f"[ ] {task['name']}")
                pending_count += 1
                
        if pending_count == 0:
            print("You have no pending tasks. Time to relax!")
        print("---------------------\n")
```

```python
# 1. Create your specific list object
my_weekend_chores = ToDoList()

#  Add some tasks 
my_weekend_chores.add_task("Buy groceries")
my_weekend_chores.add_task("Learn Python OOP")
my_weekend_chores.add_task("Call mom")

# 3. Check what you need to do
my_weekend_chores.show_pending()
```

```python
my_weekend_chores.mark_done("Buy groceries")

my_weekend_chores.show_pending()
```
